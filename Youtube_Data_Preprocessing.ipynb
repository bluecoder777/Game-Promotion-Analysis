```python
import pandas as pd
import json
import re
```

```python
# Read the file
comments_df = pd.read_csv('youtube_comments.csv')
```

```python
# Drop duplicates
comments_df = comments_df.drop_duplicates(subset= ['video_id','text','channel_id','channel_id_text']).reset_index(drop=True)
```

```python
comments_df.shape
```

```python
nan_in_columns = comments_df[comments_df[['text', 'channel_id_text']].isna().any(axis=1)]
nan_in_columns
```

```python
# Drop the NA values
cleaned_comments_df = comments_df.dropna(subset=['text', 'channel_id_text'])
```

```python
def remove_emoticons_and_url_links(text):
    # Pattern for urls
    url_pattern = r'https?://\S+|www\.\S+'

    # Pattern for emojis using unicode
    emoji_pattern = (
        r'[^\x00-\x7F]'  # Non-ASCII characters
        r'|[\U00010000-\U0010FFFF]'  # emojis
    )

    # Remove URLs
    clean_text = re.sub(url_pattern, '', text)

    # Remove emoticon
    clean_text = re.sub(emoji_pattern, '', clean_text)

    return clean_text
```

```python
cleaned_comments_df['text'] = cleaned_comments_df['text'].apply(remove_emoticons_and_url_links)
```

```python
# Write to csv
cleaned_comments_df.to_csv('cleaned_youtube_comments.csv',index = False)
```

## Suitable Youtuber Datasets

```python
def perform_cleaning(filename):
  comments_df = pd.read_csv('youtube_comments.csv')
  comments_df = comments_df.drop_duplicates(subset= ['video_id','text','channel_id','channel_id_text']).reset_index(drop=True)
  cleaned_comments_df = comments_df.dropna(subset=['text', 'channel_id_text'])
  cleaned_comments_df['text'] = cleaned_comments_df['text'].apply(remove_emoticons_and_url_links)

  cleaned_comments_df.to_csv('processed_'+filename,index = False)
```

```python
files = ['Blox4Fun_comments.csv', 'ChilledChaosGAME_comments.csv', 'Mithzan_comments.csv', 'SakawGaming_comments.csv', 'Sam Tabor Gaming_comments.csv']
```

```python
for file in files:
  perform_cleaning(file)
```
